```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
prices_df = pd.read_csv("prices_round_1_day_-2.csv", sep = ";", index_col = 1)

squid_prices = prices_df[prices_df["product"] == "SQUID_INK"]
kelp_prices= prices_df[prices_df["product"] == "KELP"]
resin_prices = prices_df[prices_df["product"] == "RAINFOREST_RESIN"]
```

```python
def get_strongest_levels(prices, val = "bid"):
    price = []
    vols = []
    print(len(prices))
    for i in range(len(prices)):
        row = prices.iloc[i]
        
        # Extract volumes and prices
        volumes = np.array([row.get(f'{val}_volume_{j}', np.nan) for j in range(1, 4)])
        prices_ = np.array([row.get(f'{val}_price_{j}', np.nan) for j in range(1, 4)])
        
        # Mask out NaNs
        mask = ~np.isnan(volumes) & ~np.isnan(prices_)
        
        if not any(mask):
            # No valid levels
            price.append(np.nan)
            vols.append(np.nan)
            continue

        valid_vols = volumes[mask]
        valid_prices = prices_[mask]

        max_index = np.argmax(valid_vols)
        price.append(valid_prices[max_index])
        vols.append(valid_vols[max_index])
    
    return price, vols
```

```python
sb, sbv = get_strongest_levels(kelp_prices, "bid")
kelp_prices["strong_bid_price"] = sb
kelp_prices["strong_bid_vol"] = sbv

sa, sav = get_strongest_levels(kelp_prices, "ask")
kelp_prices["strong_ask_price"] = sa
kelp_prices["strong_ask_vol"] = sav
```

```python

# The data appears to be malformatted, so I'll recreate it with a clearer structure
# Assuming the columns are: Index, Cluster, Symbol, Bid, BidSize, Ask, AskSize, Mid, Spread

# Sample data based on the provided information
df = pd.read_csv("squid_ink_clusters_bots.csv")
df = df.iloc[:100]
# Read the data into a DataFrame


# Create a figure
plt.figure(figsize=(10, 6))

# Plot the mid price line
plt.plot(df.index, df['mid_price'], 'b-', linewidth=2, label='Mid Price')
print('check')
# Define a color map for clusters
cluster_colors = {
    -10: 'red',
    -1100: 'green',
    -1200: 'purple',
    -1300: 'orange',
    -1400: 'cyan'
}
print('check')

# Plot markers for each point based on cluster
for idx, row in df.iterrows():
    cluster = row['cluster']
    color = cluster_colors.get(cluster, 'black')
    
    # Plot 'x' marker for each point
    plt.plot(idx, row['mid_price'], 'x', color=color, markersize=10, markeredgewidth=2, 
             label=f'Cluster {cluster}' if cluster not in [r['cluster'] for _, r in df.iloc[:idx].iterrows()] else "")
print('check')

# Customize the plot
plt.title('Mid Price with Cluster Markers', fontsize=16)
plt.xlabel('Index', fontsize=12)
plt.ylabel('Mid Price', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Add legend (only show each cluster once in legend)
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys(), loc='best')

# Show the plot
plt.tight_layout()
plt.show()

# Alternative version using scatter plot instead of 'x' markers
plt.figure(figsize=(10, 6))

# Plot the mid price line
plt.plot(df.index, df['mid_price'], 'b-', linewidth=2, label='Mid Price')
print('check')

# Add colored dots for each cluster
for idx, row in df.iterrows():
    cluster = row['cluster']
    color = cluster_colors.get(cluster, 'black')
    
    # Plot dot marker
    plt.scatter(idx, row['mid_price'], color=color, s=100, 
                label=f'Cluster {cluster}' if cluster not in [r['cluster'] for _, r in df.iloc[:idx].iterrows()] else "")
print('check')

# Customize the plot
plt.title('Mid Price with Cluster Dot Markers', fontsize=16)
plt.xlabel('Index', fontsize=12)
plt.ylabel('Mid Price', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Add legend
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys(), loc='best')

plt.tight_layout()
plt.show()
```
