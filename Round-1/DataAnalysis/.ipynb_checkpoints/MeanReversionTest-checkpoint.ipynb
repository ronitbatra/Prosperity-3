```python
import pandas as pd
import numpy as np

class BacktestStrategy:
    def __init__(self, prices, volume_threshold=5, trade_size=10):
        self.prices = prices  # DataFrame containing prices and volumes
        self.volume_threshold = volume_threshold  # Threshold for volume
        self.trade_size = trade_size  # Fixed trade size
        self.buy_prices = []  # Track buy prices
        self.sell_prices = []  # Track sell prices
        self.pnl = []  # Track profit and loss
        self.open_position = None  # Track open position for backtest
    
    def execute(self):
        # Loop through the price data and check for buy/sell conditions
        for i in range(len(self.prices)):
            # Buy condition: If ask volume at level 2 exceeds threshold
            if self.prices['ask_volume_3'].iloc[i] < self.volume_threshold and self.open_position is None:
                buy_price = self.prices['ask_price_1'].iloc[i]
                self.buy_prices.append(buy_price)
                self.open_position = buy_price  # Open a buy position
                #print(f"Buy at {buy_price} at time {self.prices.index[i]}")
                
            # Sell condition: If bid volume at level 2 exceeds threshold and there's an open position
            elif self.prices['bid_volume_3'].iloc[i] < self.volume_threshold and self.open_position is not None:
                sell_price = self.prices['bid_price_1'].iloc[i]
                self.sell_prices.append(sell_price)
                self.pnl.append(self.calculate_pnl(self.open_position, sell_price))
                self.open_position = None  # Close the position
                #print(f"Sell at {sell_price} at time {self.prices.index[i]}")
    
    def calculate_pnl(self, buy_price, sell_price):
        # Calculate the PnL for a completed trade (buy to sell)
        return (sell_price - buy_price) * self.trade_size
    
    def get_total_pnl(self):
        return np.sum(self.pnl)
    
    def plot_pnl(self):
        # Plot the PnL over time
        import matplotlib.pyplot as plt
        cumulative_pnl = np.cumsum(self.pnl)
        plt.plot(cumulative_pnl)
        plt.title('Cumulative PnL of Backtest Strategy')
        plt.xlabel('Time')
        plt.ylabel('Cumulative PnL')
        plt.show()


# Initialize the backtest strategy
backtest = BacktestStrategy(prices, volume_threshold=10, trade_size=10)

# Execute the strategy
backtest.execute()

# Calculate and print total PnL
print(f"Total PnL: {backtest.get_total_pnl()}")

# Plot the cumulative PnL
backtest.plot_pnl()
```

```python
import numpy as np
import pandas as pd

def get_strongest_levels(prices):
    price = []
    vols = []
    print(len(prices))
    for i in range(len(prices)):
        row = prices.iloc[i]
        
        # Extract volumes and prices
        volumes = np.array([row.get(f'bid_volume_{j}', np.nan) for j in range(1, 4)])
        prices_ = np.array([row.get(f'bid_price_{j}', np.nan) for j in range(1, 4)])
        
        # Mask out NaNs
        mask = ~np.isnan(volumes) & ~np.isnan(prices_)
        
        if not any(mask):
            # No valid levels
            price.append(np.nan)
            vols.append(np.nan)
            continue

        valid_vols = volumes[mask]
        valid_prices = prices_[mask]

        max_index = np.argmax(valid_vols)
        price.append(valid_prices[max_index] if valid_vols[max_index] >= 20 else price[-1])
        vols.append(valid_vols[max_index] if valid_vols[max_index] >= 20 else vols[-1])
    
    return price, vols
```

```python
# Assuming `prices_df` has a column `mid_price` and a `timestamp` column
# Replace `mid_price` with actual price calculation if needed

# Parameters
window = 10 # Rolling window size for mean and standard deviation
entry_zscore = .7  # Entry threshold for z-score
exit_zscore = 0  # Exit threshold for z-score (when z-score goes back to the mean)

# Calculate rolling mean and standard deviation
prices['rolling_mean'] = prices['new_mid'].rolling(window=window).mean()
prices['rolling_std'] = prices['new_mid'].rolling(window=window).std()

# Calculate the z-score
prices['z_score'] = (prices['new_mid'] - prices['rolling_mean']) / prices['rolling_std']
prices = prices.dropna(subset = ['z_score'])

position = 0  # 1 for long, -1 for short, 0 for no position
entry_price = 0
capital = 10000  # Starting capital for backtest
capital_curve = []  # To track the capital over time

# Backtest loop
for i in range(len(prices)):
    z_score = prices['z_score'].iloc[i]
    
    # Check for trade signals
    if position == 0:  # No current position
        if z_score < -entry_zscore:  # Buy signal
            position = 1
            entry_price = prices['ask_price_1'].iloc[i]
        elif z_score > entry_zscore:  # Sell signal
            position = -1
            entry_price = prices['bid_price_1'].iloc[i]
    elif position == 1:  # Long position
        if z_score >= exit_zscore:  # Exit long position when z-score returns to mean
            capital += (prices['bid_price_1'].iloc[i] - entry_price) * capital / entry_price
            position = 0  # Close position
    elif position == -1:  # Short position
        if z_score <= exit_zscore:  # Exit short position when z-score returns to mean
            capital += (entry_price - prices['ask_price_1'].iloc[i]) * capital / entry_price
            position = 0  # Close position
    
    # Track capital over time
    capital_curve.append(capital)

# Plot the results
prices['capital'] = capital_curve

fig, ax1 = plt.subplots(figsize=(14, 7))
ax1.plot(prices['timestamp'], prices['mid_price'], label='Mid Price', color='blue', alpha=0.6)
ax1.set_ylabel('Mid Price')
ax1.set_xlabel('Timestamp')

ax2 = ax1.twinx()
ax2.plot(prices['timestamp'], prices['capital'], label='Capital', color='green', alpha=0.6)
ax2.set_ylabel('Capital')

plt.title("Mean Reversion Backtest: Mid Price and Capital Over Time")
plt.legend(loc='best')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from collections import defaultdict

# Load data
# Assuming you have 3 order book files and 3 trade files with matching timestamps
def load_data(order_book_file, trades_file):
    """
    Load order book and trades data
    
    Args:
        order_book_file: Path to order book CSV
        trades_file: Path to trades CSV
    
    Returns:
        order_book_df: DataFrame with order book data
        trades_df: DataFrame with trades data
    """
    # Load order book data
    # Expected columns: timestamp, bid_price1, bid_size1, bid_price2, bid_size2, bid_price3, bid_size3,
    #                  ask_price1, ask_size1, ask_price2, ask_size2, ask_price3, ask_size3
    order_book_df = pd.read_csv(order_book_file)
    
    # Load trades data
    # Expected columns: timestamp, price, quantity, aggressor_side (buy/sell)
    trades_df = pd.read_csv(trades_file)
    
    # Ensure timestamps are in the same format and sorted
    order_book_df['timestamp'] = pd.to_datetime(order_book_df['timestamp'])
    trades_df['timestamp'] = pd.to_datetime(trades_df['timestamp'])
    
    order_book_df = order_book_df.sort_values('timestamp')
    trades_df = trades_df.sort_values('timestamp')
    
    return order_book_df, trades_df

def calculate_mid_price(order_book_df):
    """Calculate mid price from best bid and ask"""
    order_book_df['mid_price'] = (order_book_df['bid_price1'] + order_book_df['ask_price1']) / 2
    return order_book_df

def calculate_spread(order_book_df):
    """Calculate spread from best bid and ask"""
    order_book_df['spread'] = order_book_df['ask_price1'] - order_book_df['bid_price1']
    return order_book_df

def calculate_book_imbalance(order_book_df):
    """Calculate order book imbalance at top of book"""
    bid_volume = order_book_df['bid_size1'] + order_book_df['bid_size2'] + order_book_df['bid_size3']
    ask_volume = order_book_df['ask_size1'] + order_book_df['ask_size2'] + order_book_df['ask_size3']
    
    # Avoid division by zero
    total_volume = bid_volume + ask_volume
    order_book_df['imbalance'] = np.where(
        total_volume > 0, 
        (bid_volume - ask_volume) / total_volume,
        0
    )
    return order_book_df

def add_moving_averages(df, column='mid_price', windows=[20, 50, 200]):
    """Add moving averages to dataframe"""
    for window in windows:
        df[f'{column}_ma{window}'] = df[column].rolling(window=window).mean()
    return df

def calculate_volatility(df, column='mid_price', window=20):
    """Calculate rolling volatility"""
    df[f'{column}_volatility'] = df[column].rolling(window=window).std()
    return df

def calculate_zscore(df, column='mid_price', window=50):
    """Calculate z-score for mean reversion signals"""
    rolling_mean = df[column].rolling(window=window).mean()
    rolling_std = df[column].rolling(window=window).std()
    df[f'{column}_zscore'] = (df[column] - rolling_mean) / rolling_std
    return df

def match_trades_to_book(order_book_df, trades_df):
    """
    For each trade, find the corresponding order book state right before the trade
    and analyze if the trade was at bid, ask, or in between
    """
    merged_data = []
    
    for _, trade in trades_df.iterrows():
        # Find the latest order book state before this trade
        book_state = order_book_df[order_book_df['timestamp'] < trade['timestamp']].iloc[-1]
        
        # Determine trade type
        if trade['price'] <= book_state['bid_price1']:
            trade_type = 'at_bid'
        elif trade['price'] >= book_state['ask_price1']:
            trade_type = 'at_ask'
        else:
            trade_type = 'inside_spread'
            
        # Calculate price impact (post-trade mid price change)
        next_book = order_book_df[order_book_df['timestamp'] > trade['timestamp']].iloc[0]
        price_impact = next_book['mid_price'] - book_state['mid_price']
        
        merged_data.append({
            'trade_timestamp': trade['timestamp'],
            'book_timestamp': book_state['timestamp'],
            'mid_price': book_state['mid_price'],
            'spread': book_state['spread'],
            'trade_price': trade['price'],
            'trade_quantity': trade['quantity'],
            'trade_type': trade_type,
            'aggressor_side': trade.get('aggressor_side', None),
            'price_impact': price_impact,
            'imbalance': book_state.get('imbalance', 0)
        })
    
    return pd.DataFrame(merged_data)

def cluster_orders_by_size(order_book_df, n_clusters=3):
    """
    Cluster orders by size to identify potential bot patterns
    """
    # Extract bid orders
    bid_orders = []
    for i in range(1, 4):  # for 3 levels
        bid_sizes = order_book_df[f'bid_size{i}'].dropna()
        if not bid_sizes.empty:
            for size in bid_sizes:
                if size > 0:
                    bid_orders.append([size])
    
    # Extract ask orders
    ask_orders = []
    for i in range(1, 4):  # for 3 levels
        ask_sizes = order_book_df[f'ask_size{i}'].dropna()
        if not ask_sizes.empty:
            for size in ask_sizes:
                if size > 0:
                    ask_orders.append([size])
    
    # Combine all orders
    all_orders = np.array(bid_orders + ask_orders)
    
    # Apply K-means clustering
    if len(all_orders) > 0:
        kmeans = KMeans(n_clusters=n_clusters, random_state=42).fit(all_orders)
        centers = kmeans.cluster_centers_
        
        # Sort clusters by size
        sorted_indices = centers.flatten().argsort()
        
        # Categorize clusters
        labels = ["Small MM", "Medium MM", "Large MM"]
        if len(sorted_indices) < len(labels):
            labels = labels[:len(sorted_indices)]
        
        cluster_info = {labels[i]: centers[sorted_indices[i]][0] for i in range(len(sorted_indices))}
        return kmeans, cluster_info
    
    return None, {}

def analyze_bot_behavior_around_price_events(merged_df, order_book_df, cluster_info):
    """
    Analyze bot behavior around significant price events
    """
    # Identify significant price events
    merged_df['price_change'] = merged_df['mid_price'].diff().abs()
    threshold = merged_df['price_change'].quantile(0.95)  # Top 5% of price movements
    price_events = merged_df[merged_df['price_change'] > threshold]
    
    bot_behavior = {}
    
    for _, event in price_events.iterrows():
        event_time = event['trade_timestamp']
        
        # Get order book before and after the event
        before_window = 5  # 5 observations before
        after_window = 10  # 10 observations after
        
        idx = order_book_df[order_book_df['timestamp'] < event_time].index[-1]
        before_idx = max(0, idx - before_window)
        after_idx = min(len(order_book_df) - 1, idx + after_window)
        
        before_book = order_book_df.iloc[before_idx:idx+1]
        after_book = order_book_df.iloc[idx+1:after_idx+1]
        
        # Analyze changes in order sizes
        for bot, size in cluster_info.items():
            size_epsilon = size * 0.1  # Allow 10% variation
            
            # Count occurrences before
            before_count = count_size_occurrences(before_book, size, size_epsilon)
            
            # Count occurrences after
            after_count = count_size_occurrences(after_book, size, size_epsilon)
            
            # Record behavior
            direction = "up" if event['price_change'] > 0 else "down"
            key = (bot, direction)
            
            if key not in bot_behavior:
                bot_behavior[key] = []
                
            # Calculate behavior metrics
            response = {
                'event_time': event_time,
                'price_change': event['price_change'],
                'before_count': before_count,
                'after_count': after_count,
                'change_ratio': after_count / before_count if before_count > 0 else float('inf')
            }
            
            bot_behavior[key].append(response)
    
    # Summarize bot behavior
    behavior_summary = {}
    
    for key, responses in bot_behavior.items():
        bot, direction = key
        avg_change_ratio = np.mean([r['change_ratio'] for r in responses if not np.isinf(r['change_ratio'])])
        behavior_summary[key] = {
            'avg_change_ratio': avg_change_ratio,
            'count': len(responses)
        }
    
    return price_events, behavior_summary

def count_size_occurrences(book_df, target_size, epsilon):
    """Count occurrences of orders with sizes close to target_size"""
    count = 0
    
    # Check bid sizes
    for i in range(1, 4):
        sizes = book_df[f'bid_size{i}']
        count += ((sizes >= target_size - epsilon) & (sizes <= target_size + epsilon)).sum()
    
    # Check ask sizes
    for i in range(1, 4):
        sizes = book_df[f'ask_size{i}']
        count += ((sizes >= target_size - epsilon) & (sizes <= target_size + epsilon)).sum()
    
    return count

def analyze_reversion_patterns(merged_df):
    """
    Analyze price reversion patterns after large moves
    """
    # Calculate future returns over different horizons
    for horizon in [1, 3, 5, 10]:
        merged_df[f'future_return_{horizon}'] = merged_df['mid_price'].pct_change(periods=horizon).shift(-horizon)
    
    # Group z-scores into buckets
    merged_df['zscore_bucket'] = pd.cut(merged_df['mid_price_zscore'], 
                                         bins=[-np.inf, -3, -2, -1, 0, 1, 2, 3, np.inf],
                                         labels=[-3, -2, -1, 0, 1, 2, 3])
    
    # Calculate average future returns by z-score bucket
    reversion_stats = merged_df.groupby('zscore_bucket')[
        ['future_return_1', 'future_return_3', 'future_return_5', 'future_return_10']
    ].mean()
    
    return reversion_stats

def create_trading_signals(merged_df, zscore_threshold=2.0, vol_filter=True):
    """
    Create trading signals based on mean reversion and volatility filter
    
    Args:
        merged_df: DataFrame with merged order book and trade data
        zscore_threshold: Z-score threshold for generating signals
        vol_filter: Whether to apply volatility filter
        
    Returns:
        merged_df with additional signal columns
    """
    # Base mean reversion signals
    merged_df['long_signal'] = (merged_df['mid_price_zscore'] < -zscore_threshold)
    merged_df['short_signal'] = (merged_df['mid_price_zscore'] > zscore_threshold)
    
    # Apply volatility filter if requested
    if vol_filter:
        vol_percentile = merged_df['mid_price_volatility'].rolling(window=100).rank(pct=True)
        high_vol = (vol_percentile > 0.7)  # Top 30% of volatility
        merged_df['long_signal'] = merged_df['long_signal'] & high_vol
        merged_df['short_signal'] = merged_df['short_signal'] & high_vol
    
    # Add imbalance filter - stronger signal if imbalance agrees with reversal direction
    merged_df['long_signal_strong'] = merged_df['long_signal'] & (merged_df['imbalance'] > 0.3)
    merged_df['short_signal_strong'] = merged_df['short_signal'] & (merged_df['imbalance'] < -0.3)
    
    return merged_df

def execute_analysis(order_book_file, trades_file):
    """
    Main function to execute the full analysis pipeline
    """
    # Load data
    order_book_df, trades_df = load_data(order_book_file, trades_file)
    
    # Process order book
    order_book_df = calculate_mid_price(order_book_df)
    order_book_df = calculate_spread(order_book_df)
    order_book_df = calculate_book_imbalance(order_book_df)
    order_book_df = add_moving_averages(order_book_df)
    order_book_df = calculate_volatility(order_book_df)
    order_book_df = calculate_zscore(order_book_df)
    
    # Match trades to order book states
    merged_df = match_trades_to_book(order_book_df, trades_df)
    
    # Propagate derived metrics from order book to merged dataframe
    for col in ['mid_price_ma20', 'mid_price_ma50', 'mid_price_ma200', 
                'mid_price_volatility', 'mid_price_zscore']:
        if col in order_book_df.columns:
            merged_df = pd.merge_asof(
                merged_df.sort_values('trade_timestamp'),
                order_book_df[['timestamp', col]].sort_values('timestamp'),
                left_on='trade_timestamp',
                right_on='timestamp',
                direction='backward'
            )
    
    # Cluster orders by size to identify bots
    kmeans, cluster_info = cluster_orders_by_size(order_book_df)
    
    # Analyze bot behavior around price events
    price_events, bot_behavior = analyze_bot_behavior_around_price_events(
        merged_df, order_book_df, cluster_info
    )
    
    # Analyze reversion patterns
    reversion_stats = analyze_reversion_patterns(merged_df)
    
    # Create trading signals
    merged_df = create_trading_signals(merged_df)
    
    return {
        'order_book_df': order_book_df,
        'trades_df': trades_df,
        'merged_df': merged_df,
        'cluster_info': cluster_info,
        'price_events': price_events,
        'bot_behavior': bot_behavior,
        'reversion_stats': reversion_stats
    }

def visualize_results(results):
    """
    Create visualizations of the analysis results
    """
    order_book_df = results['order_book_df']
    merged_df = results['merged_df']
    reversion_stats = results['reversion_stats']
    cluster_info = results['cluster_info']
    
    # Create figure with multiple subplots
    fig = plt.figure(figsize=(20, 15))
    
    # 1. Price and moving averages
    ax1 = fig.add_subplot(3, 2, 1)
    ax1.plot(order_book_df['timestamp'], order_book_df['mid_price'], label='Mid Price')
    if 'mid_price_ma20' in order_book_df.columns:
        ax1.plot(order_book_df['timestamp'], order_book_df['mid_price_ma20'], label='20 MA')
    if 'mid_price_ma50' in order_book_df.columns:
        ax1.plot(order_book_df['timestamp'], order_book_df['mid_price_ma50'], label='50 MA')
    ax1.set_title('Squid Ink Price and Moving Averages')
    ax1.legend()
    plt.xticks(rotation=45)
    
    # 2. Z-score and significant deviations
    ax2 = fig.add_subplot(3, 2, 2)
    if 'mid_price_zscore' in order_book_df.columns:
        ax2.plot(order_book_df['timestamp'], order_book_df['mid_price_zscore'])
        ax2.axhline(y=2, color='r', linestyle='--')
        ax2.axhline(y=-2, color='r', linestyle='--')
        ax2.set_title('Price Z-score (Deviation from 50-period Mean)')
    plt.xticks(rotation=45)
    
    # 3. Bot order size clusters
    ax3 = fig.add_subplot(3, 2, 3)
    
    # Flatten and collect all order sizes for visualization
    all_sizes = []
    for i in range(1, 4):
        all_sizes.extend(order_book_df[f'bid_size{i}'].dropna().tolist())
        all_sizes.extend(order_book_df[f'ask_size{i}'].dropna().tolist())
    
    all_sizes = [s for s in all_sizes if s > 0]
    
    if all_sizes:
        sns.histplot(all_sizes, kde=True, ax=ax3)
        
        # Add vertical lines for cluster centers
        for bot, size in cluster_info.items():
            ax3.axvline(x=size, color='r', linestyle='--', label=f"{bot}: {size:.1f}")
        
        ax3.set_title('Order Size Distribution and Bot Clusters')
        ax3.legend()
    
    # 4. Mean reversion analysis
    ax4 = fig.add_subplot(3, 2, 4)
    if not reversion_stats.empty:
        reversion_stats.plot(kind='bar', ax=ax4)
        ax4.set_title('Average Returns After Z-score Levels')
        ax4.set_ylabel('Average Return')
        ax4.set_xlabel('Z-score Bucket')
    
    # 5. Spread over time
    ax5 = fig.add_subplot(3, 2, 5)
    ax5.plot(order_book_df['timestamp'], order_book_df['spread'])
    ax5.set_title('Bid-Ask Spread Over Time')
    plt.xticks(rotation=45)
    
    # 6. Trading signals
    ax6 = fig.add_subplot(3, 2, 6)
    ax6.plot(merged_df['trade_timestamp'], merged_df['mid_price'], label='Mid Price')
    
    # Plot long and short signals
    long_signals = merged_df[merged_df['long_signal']]
    short_signals = merged_df[merged_df['short_signal']]
    
    if not long_signals.empty:
        ax6.scatter(long_signals['trade_timestamp'], 
                   long_signals['mid_price'], 
                   color='green', marker='^', label='Long Signal')
    
    if not short_signals.empty:
        ax6.scatter(short_signals['trade_timestamp'], 
                   short_signals['mid_price'], 
                   color='red', marker='v', label='Short Signal')
    
    ax6.set_title('Trading Signals Based on Mean Reversion')
    ax6.legend()
    plt.xticks(rotation=45)
    
    plt.tight_layout()
    plt.show()
    
    return fig

# Example usage:
# results = execute_analysis('order_book_data.csv', 'trades_data.csv')
# visualize_results(results)

# Function to backtest the trading strategy
def backtest_strategy(merged_df, initial_capital=10000, position_size=0.2):
    """
    Backtest a simple mean reversion trading strategy
    
    Args:
        merged_df: DataFrame with merged order book and trade data
        initial_capital: Initial capital for backtesting
        position_size: Fraction of capital to use per trade
        
    Returns:
        DataFrame with backtest results
    """
    backtest_df = merged_df.copy()
    
    backtest_df['position'] = 0
    backtest_df['position'] = np.where(backtest_df['long_signal'], 1, backtest_df['position'])
    backtest_df['position'] = np.where(backtest_df['short_signal'], -1, backtest_df['position'])
    
    # Calculate pnl (using next period's return)
    backtest_df['return'] = backtest_df['mid_price'].pct_change().shift(-1)
    backtest_df['strategy_return'] = backtest_df['position'] * backtest_df['return']
    
    # Calculate cumulative returns
    backtest_df['cum_return'] = (1 + backtest_df['return']).cumprod() - 1
    backtest_df['cum_strategy_return'] = (1 + backtest_df['strategy_return']).cumprod() - 1
    
    # Calculate equity curve
    backtest_df['equity'] = initial_capital * (1 + backtest_df['cum_strategy_return'])
    
    # Calculate trade stats
    trades = backtest_df[backtest_df['position'] != backtest_df['position'].shift(1)].copy()
    trades['trade_type'] = np.where(trades['position'] == 1, 'Long', 
                                    np.where(trades['position'] == -1, 'Short', 'Flat'))
    
    # Calculate sharp ratio
    if len(backtest_df) > 0:
        annual_factor = 252 * 24  # Assuming hourly data for high-frequency trading
        sharpe = (backtest_df['strategy_return'].mean() / backtest_df['strategy_return'].std() * 
                  np.sqrt(annual_factor)) if backtest_df['strategy_return'].std() > 0 else 0
    else:
        sharpe = 0
    
    return backtest_df, trades, sharpe

# Function to optimize strategy parameters
def optimize_strategy(merged_df, zscore_thresholds=[1.5, 2.0, 2.5, 3.0], 
                     vol_filters=[True, False]):
    """
    Optimize strategy parameters
    
    Args:
        merged_df: DataFrame with merged order book and trade data
        zscore_thresholds: List of z-score thresholds to test
        vol_filters: List of boolean values for applying volatility filter
        
    Returns:
        DataFrame with optimization results
    """
    results = []
    
    for zscore in zscore_thresholds:
        for vol_filter in vol_filters:
            # Create signals with current parameters
            test_df = create_trading_signals(merged_df.copy(), 
                                            zscore_threshold=zscore, 
                                            vol_filter=vol_filter)
            
            # Backtest
            backtest_df, trades, sharpe = backtest_strategy(test_df)
            
            # Calculate additional metrics
            win_rate = (backtest_df['strategy_return'] > 0).mean() if len(backtest_df) > 0 else 0
            final_return = backtest_df['cum_strategy_return'].iloc[-1] if len(backtest_df) > 0 else 0
            max_drawdown = (backtest_df['equity'] / backtest_df['equity'].cummax() - 1).min() if len(backtest_df) > 0 else 0
            
            results.append({
                'zscore_threshold': zscore,
                'vol_filter': vol_filter,
                'sharpe': sharpe,
                'win_rate': win_rate,
                'final_return': final_return,
                'max_drawdown': max_drawdown,
                'trade_count': len(trades)
            })
    
    return pd.DataFrame(results)
```
